```python
with open ("/content/drive/MyDrive/Cricket_Analytics_Project/t20_json_files/t20_json_files/t20_wc_match_results.json") as f:
  data=json.load(f)
df_match=pd.DataFrame(data[0]["matchSummary"])
df_match.head()
```

```python
df_match.rename({"scorecard":"id"},axis=1,inplace=True)
```

```python
matchids={}
for index,row in df_match.iterrows():
  key1 = row['team1'] + " Vs " + row['team2']
  key2 = row["team2"] + " Vs " + row["team1"]

  matchids[key1]=row["id"]
  matchids[key2]=row["id"]

matchids

```

```python
df_match.to_csv("/content/drive/MyDrive/Cricket_Analytics_Project/csv files/matchsummary.csv")
```

BATTING **PREPROCESSINGS**

```python
with open("/content/drive/MyDrive/Cricket_Analytics_Project/t20_json_files/t20_json_files/t20_wc_batting_summary.json") as f:
  data=json.load(f)

  all_matches=[]
  for rec in data:
    all_matches.extend(rec["battingSummary"])

df_batting=pd.DataFrame(all_matches)
df_batting
```

```python
df_batting["Out/N_out"]=df_batting.dismissal.apply(lambda x:"Out" if len(x)>0 else "Notout" )
df_batting.drop(columns=["dismissal"],inplace=True)
```

```python
df_batting["batsmanName"]=df_batting.batsmanName.apply(lambda x: x.replace("†",""))
```

```python
df_batting["id"]=df_batting.match.map(matchids)
df_batting.head()
```

```python
df_batting.to_csv('/content/drive/MyDrive/Cricket_Analytics_Project/csv files/batting.csv', index = False)
```
